```python
#ans-3  FloatingPointError, OverflowError, ZeroDivisionError, DecimalException
#are defined in ArithmeticError Class. Example are-

try:
    5/0
except ZeroDivisionError as e:
    print(e)
    
try:
    j=5.0
    for i in range(1,1000):
        j=j**i
        print(j)
except OverflowError as e:
    print(e)
```

```python
#Ans-4 LookupError class raised an exception when an index or a 
#key is not found for a sequence or dictionary.
#Example of key error-
try:
    d= {"key": "sudh", 1: [2,3,4,5]}
    print(d["key2"])
except KeyError as e:
    print(e)
    
#Example of index error-
try:
    l= [1,2,3,"sudh",45]
    print(l[7])
except IndexError as e:
    print(e)
```
